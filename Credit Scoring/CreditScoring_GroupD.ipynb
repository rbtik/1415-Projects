#### TESTS OF COMPARISON

Mann-Whitney-Wilcoxon (MWW) RankSum test.

The MWW RankSum test is a useful test to determine if two distributions are 
significantly different or not. Unlike the t-test, the RankSum test does not assume that the data are normally distributed, potentially providing a more accurate assessment of the data sets.

```python
#Comparison of Age by Creditability i.e. Goods/Bads
from scipy import stats  
credit_goods = credit2[credit2['creditability'] == 1]['age']
credit_bads = credit2[credit2['creditability'] == 0]['age']
z_stat, p_val = stats.ranksums(credit_goods, credit_bads)
print "P value =", p_val
#The Average Age of 'Goods' is significantly different from 'Bad'
```

```python
#Comparison of Savings by Creditability i.e. Goods/Bads
credit_goods_sav = credit2[credit2['creditability'] == 1]['savings_or_stocks']
credit_bads_sav = credit2[credit2['creditability'] == 0]['savings_or_stocks']
z_stat, p_val = stats.ranksums(credit_goods_sav, credit_bads_sav)
print "P value =", p_val
#The average savings for goods is significantly different from bads
```

```python
#Comparison of Credit in use  by Creditability i.e. Goods/Bads
credit_goods_creduse = credit2[credit2['creditability'] == 1]['credit_in_use']
credit_bads_creduse = credit2[credit2['creditability'] == 0]['credit_in_use']
z_stat, p_val = stats.ranksums(credit_goods_creduse, credit_bads_creduse)
print "P value =", p_val
#The average credit in use for goods is significantly different from bads
```

```python
#Comparison of economic dependants  by Creditability i.e. for Goods/Bads
credit_goods_dep = credit2[credit2['creditability'] == 1]['economic_dependants']
credit_bads_dep = credit2[credit2['creditability'] == 0]['economic_dependants']
z_stat, p_val = stats.ranksums(credit_goods_dep, credit_bads_dep)
print "P value =", p_val
#There is not much difference between goods and bads in terms of number of economic dependants
```

```python
#Comparison of economic dependants  by Creditability i.e. for Goods/Bads
credit_goods_dep = credit2[credit2['creditability'] == 1]['duration']
credit_bads_dep = credit2[credit2['creditability'] == 0]['duration']
z_stat, p_val = stats.ranksums(credit_goods_dep, credit_bads_dep)
print "P value =", p_val
#There is not much difference between goods and bads in terms of number of economic dependants
```

### Model Development

```python
#Checking Correlations
credit2.corr()
```

We want to add an intercept column as well as create some dummy variables for categorical variables. 
The dmatrices function from the patsy module can do that using formula language.

```python
import statsmodels.api as sm
from patsy import dmatrices
from sklearn.linear_model import LogisticRegression
from sklearn.cross_validation import train_test_split
from sklearn import metrics
from sklearn.cross_validation import cross_val_score

y, X = dmatrices('creditability ~ duration + credit_in_use + savings_or_stocks + age + curr_hhld_tenure + C(balance_credit_acc) +C(moral) + C(credit_purpose) + C(years_curr_employer) + C(rate) + C(status_sex) + C(debtors_guarantors) + \
                  C(assets) + C(further_credits) + C(apartment_type) + prev_credits_here + C(occupation) + economic_dependants + C(telephone) + C(foreign_worker)',credit2, return_type="dataframe")
print X.columns 

```

```python
# flatten y into a 1-D array
y = np.ravel(y)
```

```python
# instantiate a logistic regression model, and fit with X and y
model = LogisticRegression()
model = model.fit(X, y)

# check the accuracy on the training set
model.score(X, y)
```

```python
#Dropping Age for a Model Comparison(by including ageGroups instead of Age)
credit3=credit2.drop(['age'], axis=1)
```

#### Getting Paramater Estimate Table for understanding Significant Variables in the Model

```python
# We are using Logit Function from StatsModels to build the Model.
#Creditability is the dependent Variable in the below code chunk.

train_cols = credit2.columns[1:]
 
logit = sm.Logit(credit2['creditability'], credit2[train_cols])
 
# fit the model
result = logit.fit()
print result.summary()
```

The variables that were found significant in the Model (at 95% Confidence Level)and that increase or decrease the likelihood of a customer being classified as 'GOOD' were as follows:

Balance Credit Account-> Higher balance increases the likelihood of customer being classified as'Good'

Duration of Credit-> Longer a customer takes to pay back-decrease the likelihood of customer being classified as 'Good'

Moral-Payment of Previous Credits-> If a customer has regularly cleared his past credits-it increase his likelihood of being classified as 'Good'

Savings or Stocks-> More savings increase the likelihood of customer being classified as 'Good'

Rate-> (Ratio of Installment to Income)-> Higher Rate, decreases the likelihood of customer being classified as 'Good'

Assets-> More the number of assets, lesser the likelihood of being 'Good'

Number of Previous Credits at the Bank-> Higher the number of previous credits at the Bank, lesser the likelihood of being 'Good'.



```python
#Running Logit by including ageGroups instead of age to understand if bucketing of age plays a role in increasing its significance
#Getting Paramater Estimate Table for understanding Significant Variables in the Model (Model with ageGroups)
train_cols = credit3.columns[1:]
 
logit = sm.Logit(credit3['creditability'], credit3[train_cols])
 
# fit the model
result = logit.fit()
print result.summary()
```

Note on Age vs. AgeGroups: It is interesting to note that the P Value for ageGroups is 0.107, but it still is not significant at 95% Level, though its more 
siginificant as compared to the Age Variable
But worth noting that sometimes there is a linear relationship of a particular category of a variable with the dependent variable
which is sometimes not captured if that variable is continuous

### Model Evaluation Using a Validation Set

Split the data into a training set and a testing set.

```python
#Evaluate the model by splitting into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
model2 = LogisticRegression()
model2.fit(X_train, y_train)
```

We now need to predict class labels for the test set. We will also generate the class probabilities, just to take a look


```python
# predict class labels for the test set
predicted = model2.predict(X_test)
print predicted
```

```python
# generate class probabilities
probs = model2.predict_proba(X_test)
print probs
```

```python
#As you can see, the classifier is predicting a 1 (Good Account) any time the probability in the second column is greater than 0.5.
#In next step we need to generate some evaluation metrics.
# generate evaluation metrics

print metrics.accuracy_score(y_test, predicted)
print metrics.roc_auc_score(y_test, predicted)
```

The accuracy is 77%, which is almost the same as we experienced when training and predicting on the same data.


### Model Evaluation Using Cross-Validation 10-fold cross-validation, to see if the accuracy holds up more rigorously.

```python
# evaluate the model using 10-fold cross-validation
scores = cross_val_score(LogisticRegression(), X, y, scoring='accuracy', cv=10)
print scores
print scores.mean()
```

The Model Performance is quite good It's still performing at 70%+ accuracy and therefore can be recommended to solve this case.

# Road Clustering

## Load necessary packages and CSV files

```python
import pandas as pd
import numpy as np
from scipy.spatial.distance import pdist, squareform
from scipy.cluster.hierarchy import linkage, dendrogram
from scipy.cluster.vq import kmeans, vq
import scipy.cluster.hierarchy as hac
import matplotlib.pyplot as plt
from pandas import Series, DataFrame
from collections import Counter
import glob  
import csv
from pylab import plot,show
%matplotlib inline

# create empty lists to load data into
dataLists = [] 
pairs = []
  
# read data  
allfiles = glob.glob("*.csv")
for files in allfiles:
    df=pd.read_csv(files, sep=";", header = 0)
    df.columns=['TimeOfDay', 'AntennaSeconds']


    dataLists.append(df) # load dataframes as lists
    pairs.append(files) # load file names

final = pd.DataFrame(pairs) # create dataframe with file names. Will be used to load cluster assignments into.
```

## Plot each CSV file to take a look at the different patterns

```python
i = 0

for df in dataLists:
    plt.figure(figsize=(5,3))
    plt.scatter(df['TimeOfDay'], df['AntennaSeconds'], s=1)
    plt.title(pairs[i])
    i+=1
```

## Plot each CSV file as boxplot to take a look at how the data is distributed.

#### Boxplots are a great way to understand how data is distributed. When compared against the scatterplots of the data, patterns start to emerge. Roads tend to have lower means as well as a tighter range between the first quartile and second quartile vs the second quartile and third quartile.

```python
totfigs=len(allfiles)
plt.figure(figsize=(20,20))
plt.hold=True
boxes=[]
labels=list(pairs)
for df in dataLists:
    boxes.append(np.array(df['AntennaSeconds']))

plt.boxplot(boxes, vert=0, labels=labels)
plt.show()    
```

## Create variables to describe each CSV file in order to create a single 50 row dataset in order to better cluster the antenna pairs.

Variables created are the mean of the data for each pair of antennas and the range between the mean and the first quartile. Hypothetically, if the range is tigher (a smaller number) then the data is left skewed which is what we see with the road graphs. Moreover, the road data is not only left-skewed but also seems to have the mean and median closer to the lower end of the "time between antenna" range. As such, mean was included because range only describes tightness of data, but does not describe where the data sits on the scale, so mean was used to anchor that. 

```python
meannumber=[]
fqrange=[]

# Find mean for each dataset
for df in dataLists:
    meannumber.append((df['AntennaSeconds']).mean())

# Find the first quartile
for df in dataLists:
    fq = np.array(df['AntennaSeconds'])
    p=np.percentile(fq,25)
    fqrange.append(p)

# Find the range between the mean and the first quartile. 
#The lower the number, the tighter the distribution in the lower end of the graph.    
from operator import sub
c = map(sub, meannumber,fqrange)        

# Create 50 row lists with the variables
final['AntennaPair']=pairs
final['mean'] = meannumber
final['fqrange'] = c

#Put the lists together into the dataframe
final1 = final.drop(0,1)
final2=final1.set_index('AntennaPair')
```

# First Attempt: Use Hierarchical Clustering to try to find clusters without having to specify the number of clusters.

```python
fig, axes23 = plt.subplots(2, 3)

a = np.array(final2)

for method, axes in zip(['single', 'complete'], axes23):
    z = hac.linkage(a, method=method)

    # Plotting
    axes[0].plot(range(1, len(z)+1), z[::-1, 2])
    knee = np.diff(z[::-1, 2], 2)
    axes[0].plot(range(2, len(z)), knee)

    num_clust1 = knee.argmax() + 2
    knee[knee.argmax()] = 0
    num_clust2 = knee.argmax() + 2

    axes[0].text(num_clust1, z[::-1, 2][num_clust1-1], 'possible\n<- knee point')

    part1 = hac.fcluster(z, num_clust1, 'maxclust')
    part2 = hac.fcluster(z, num_clust2, 'maxclust')

           
    clr = ['#2200CC' ,'#D9007E' ,'#FF6600' ,'#FFCC00' ,'#ACE600' ,'#0099CC' ,
    '#8900CC' ,'#FF0000' ,'#FF9900' ,'#FFFF00' ,'#00CC01' ,'#0055CC']

    for part, ax in zip([part1, part2], axes[1:]):
        for cluster in set(part):
            ax.scatter(a[part == cluster, 0], a[part == cluster, 1], 
                       color=clr[cluster])

    m = '\n(method: {})'.format(method)
    plt.setp(axes[0], title='Knee Plot{}'.format(m), xlabel='partition',
             ylabel='{}\ncluster distance'.format(m))
    plt.setp(axes[1], title='{} Clusters'.format(num_clust1))
    plt.setp(axes[2], title='{} Clusters'.format(num_clust2))

# Print the assigned cluster numbers for the 2-cluster graph
print part1, Counter(part1)

# Plot the graphs
plt.tight_layout()
plt.show()
```

```python
data_dist = pdist(a)
data_link=linkage(data_dist)

dendrogram(data_link, labels=a.dtype.names)
plt.xlabel('Samples')
plt.ylabel('Distance')
plt.suptitle('Samples Hierarchical Clustering', fontweight='bold', fontsize=14)
```

```python
# Compute and plot first dendrogram.
fig = plt.figure(figsize=(8,8))
# x ywidth height
ax1 = fig.add_axes([0.05,0.1,0.2,0.6])
Y = linkage(data_dist, method='complete')
Z1 = dendrogram(Y, orientation='right',labels=a.dtype.names) # adding/removing the axes
ax1.set_xticks([])


# Compute and plot second dendrogram.
ax2 = fig.add_axes([0.3,0.71,0.6,0.2])
Z2 = dendrogram(Y)
ax2.set_xticks([])
ax2.set_yticks([])

#Compute and plot the heatmap
axmatrix = fig.add_axes([0.3,0.1,0.6,0.6])
idx1 = Z1['leaves']
idx2 = Z2['leaves']
D = squareform(data_dist)
D = D[idx1,:]
D = D[:,idx2]
im = axmatrix.matshow(D, aspect='auto', origin='lower', cmap=plt.cm.YlGnBu)
axmatrix.set_xticks([])
axmatrix.set_yticks([])

# Plot colorbar.
axcolor = fig.add_axes([0.91,0.1,0.02,0.6])
plt.colorbar(im, cax=axcolor)
```

The heatmap shows that there is strong yellow along the diagonal but it appears that there are smaller subclusters within the two larger clusters.

Hierarchical Clustering based on average and range of the first quartile to second quartile resulted in two clusters of 24 and 26 pairs of antennas. As it was stated in class that there are 25 roads, this is not the optimal solution.

# Second Attempt: K-Means

## Create variables to describe each CSV files in order to create a single 50 row dataset in order to better cluster the antenna pairs. 

The variables are the same as in attempt one except they were only calculated on the data after 750 minutes (around 12:30pm) giving enough time for morning activities to stabilize. This is based purely on assumption. For a better model, a more expert opinion would be needed to identify what an actual road is and to consult on what would be a representative time frame or if time is even a factor in this model.

```python
meannumber=[]
fqrange=[]
dataLists1 =[]
dataLists2=[]

#set timeframe for after the morning (12:30pm)
for df in dataLists:
    df=df[df['TimeOfDay']>750] 
    
    dataLists1.append(df)
dataLists=dataLists1

#Find mean of each dataset
for df in dataLists:
    meannumber.append((df['AntennaSeconds']).mean())

#Find Q1 for each dataset
for df in dataLists:
    fq = np.array(df['AntennaSeconds'])
    p=np.percentile(fq,25)
    fqrange.append(p)

#Find the range from mean to Q1    
from operator import sub
c = map(sub, meannumber,fqrange)        

#Create dataframe
final['AntennaPair']=pairs
final['mean'] = meannumber
final['fqrange'] = c

final1 = final.drop(0,1)
final4=final1.set_index('AntennaPair')
final4.head()
```

## Use K-Means to divide data into 2 clusters (ideally roads and non-roads)

```python
import matplotlib
#matplotlib.use('Agg')
from scipy.cluster.vq import *
import pylab
pylab.close()

b = np.array(final4) 

#computing kmeans with k=2 (2 clusters)
centroids,_=kmeans(b,2)
#assign each sample to a cluster
idx,_=vq(b,centroids)

print idx, Counter(idx)

plot(b[idx==0,0],b[idx==0,1],'ob', b[idx==1,0],b[idx==1,1], 'or')
plot(centroids[:,0],centroids[:,1],'sg',markersize=8)
show()
```

## Assign cluster number to each pair of antennas

1 is Road, 0 is Not Road

```python
#Take a look at dataframe with clustering assignments
final4['cluster']=idx 
final4.head()
```
